## Data Selection in Series

A **series** object acts in many ways like a one dimensional Numpy array and in many ways like a standard python dictionary

### Series as Dictionary

The **series** provide a mapping from a collection of keys to a collection of values

```python
import pandas as pd

data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data
```

```python
data['b']
```

```python
'a' in data
```

```python
data.keys()
```

```python
list(data.items())
```

```python
## series objects can also be modified with a dictionary-like syntax

data['e'] = 1.25
data
```

## Series as One-Dimensional Array

A **Series** builds on this dictionary like interface and provide array style item selection via the same basic mechanism as Numpy array-that is slices, masking and fancing indexing

```python
# slicing by explicit indexing
data['a':'d']
```

```python
# slicing by implicit indexing
data[0:2]
```

```python
# masking
data[(data > 0.3) & (data < 1.0)]
```

```python
# fancy indexing
data[['a', 'e']]
```

## Indexers: Loc & Iloc

Pandas provides some special **indexer** attributes expose certain indexing schemas

```python
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
data
```

```python
# explicit indexing
data[1]
```

```python
# implicit indexing
data[1:3]
```

The **Loc** attribute allows indexing and slicing that always refrencing the explicit index

```python
data.loc[1]
```

```python
data.loc[1:3]
```

The **iloc** attributes allows indexing and slicing that always refrences the implicit python style index

```python
data.iloc[1]
```

```python
data.iloc[1:3]
```

## Data Selection in DataFrames

## DataFrame as Dictionary

We will consider **Dataframe** as a dictionary of related Series objects.

```python
area = pd.Series({'California': 423967, 'Texas': 695662, 'Florida': 170312, 'New York': 141297, 'Pennsylvenia': 119280})
pop = pd.Series({'California': 39538223, 'Texas': 29145505, 'Florida': 21538187, 'New York': 20201249, 'Pennsylvenia': 13002700})

data = pd.DataFrame({'area': area, 'pop': pop})
data
```

```python
data['area']
```

```python
data.area
```

```python
data['density'] = data['pop'] / data['area']
data
```

To avoid temptation to try column assignment via attributes (use **data['pop'] = z** rather than **data.pop = z**)

## Dataframe as Two Dimensional Array

```python
data.values
```

```python
data.T # transpose row and column
```

```python
data.values[0]
```

```python
data.iloc[:3, :2]
```

```python
data.loc[:'Florida', :'pop']
```

```python
data[data.density > 120]
```
