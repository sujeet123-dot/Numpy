## Trade Offs in Missing Data Converntions

To track the presence of missing data in a table or dataframes, they generally revolve around one of two strategy: 1) **mask** that globally missing values , 2) **sentinal values** that indicates a missing entry


Pandas has two modes of storing and manipulating null values:-

1) The default mode is to use a sentinel based missing data scheme with sentinel values **NAN OR NONE** depanding on the type of the data.

2) we can opt in to using the nullable data tupes pandas providing which results in the creation an accompanied mask array to track missing values

```python
import pandas as pd
import numpy as np
```

## NaN: Missing Numerical Data

NaN is different, it is a special floating point value recognized by all systems that use the standard IEEE floating point representation.

```python
vals2 = np.array([1, np.nan, 2, 3, 4])
vals2

# Nan is a bit like data virus it infects any other bjects it touches
```

```python
1 + np.nan
```

```python
vals2.sum(), vals2.min(), vals2.max()
```

```python
np.nansum(vals2), np.nanmin(vals2), np.nanmax(vals2)
```

## Pandas Nullable Dtypes

Previously there is no way to represent a true integer array with missing data.

Pandas added **nullable dtypes**  (eg:- pd.Int32)

```python
pd.Series([1, np.nan, 2, None, pd.NA], dtype='Int32')
```

## Operation on Null Values

1) **isnull** :- Generates a boolean mask indicating missing values

2) **notnull**:- opposite of isnull

3) **dropna** :- Returned a filtered version of data

4) **fillna** :- Return a copy of the data with missing values filled or imputed

```python
data = pd.Series([1, np.nan, 'hello', None])
```

```python
data.isnull()
```

```python
data[data.isnull()]
```

```python
data[data.notnull()]
```

```python
data.dropna()
```

```python
df = pd.DataFrame([[1, np.nan, 2],
                   [2, 3, 5],
                   [np.nan, 4, 6]])
df
```

```python
df.dropna()
## we cannot drop single value in dataframe
```

```python
df.dropna(axis='columns')
```

```python
df[3] = np.nan
df
```

```python
df.dropna(axis='columns', how='all')
## how='all' only drop rows and colums that contains null values
```

```python
data = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'), dtype='Int32')
data
```

```python
data.fillna(0)
## fill NA entries with a single value such as zero
```

```python
# forward fill
data.fillna(method='ffill')
```

```python
# back fill
data.fillna(method='bfill')
```
