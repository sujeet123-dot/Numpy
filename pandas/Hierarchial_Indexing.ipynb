```python
import numpy as np
import pandas as pd
```

## A Multiply Indexed Series


```python
index = [('california', 2010), ('california', 2020), ('New York', 2010), ('New York', 2020), ('Texas', 2010), ('Texas', 2020)]
populations = [37253956, 39538223, 19378102, 20201249, 25145561, 29145505]
pop = pd.Series(populations, index=index)
pop
```

```python
pop[[i for i in pop.index if i[1] == 2010]]
```

## The Better Way: The Pandas MultiIndex

```python
index = pd.MultiIndex.from_tuples(index)
# Multiindex represents multiple level of indexing
```

```python
pop = pop.reindex(index)
pop
```

```python
pop[:, 2020]
```

## MultiIndex as Extra Dimension

The **unstack** method will quickly convert a multiply indexed series into a conventionally indexed Dataframes

The **stack** method provide the opposite operations

```python
pop_df = pop.unstack()
pop_df
```

```python
pop_df.stack()
```

```python
pop_df = pd.DataFrame({'total': pop,
                       'Under 18': [9284094, 8898092,
                                    4318033, 4181528,
                                    6879014, 7432474]})
pop_df
```

```python
f_u18 = pop_df['Under 18'] / pop_df['total']
f_u18.unstack()
```

## Method of Multiindex Creation

The most straightforward way to create a multiply indexed series or dataframes is to simply pass a list of two or more index arrays to the constructor

```python
df = pd.DataFrame(np.random.rand(4,2),
                  index=[['a','a', 'b', 'b'], [1,2,1,2]],
                  columns=['data1', 'data2'])
df
```

## MultiIndex for columns

In Dataframes, row and columns are completely symmetric

```python
# hierarchial indices and columns
index = pd.MultiIndex.from_product([[2013, 2014], [1, 2]],
                                   names=['year', 'visit'])
columns = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp']],
                                     names=['subject', 'type'])

data = np.round(np.random.randn(4, 6), 1)
data[:, ::2] *= 10
data += 37

health_data = pd.DataFrame(data=data, index=index, columns=columns)
health_data
```

```python
health_data['Guido']
```

## Rearranging Multi-Indexes

### Sorted and Unsorted Indices

Many of the MultiIndex slicing will fail if the index is not sorted.

```python
index = pd.MultiIndex.from_product([['a', 'b', 'c'], [1,2]])
data = pd.Series(np.random.rand(6), index=index)
data.index.names = ['char', 'int']
data
```

```python
try:
  data['a': 'b']
except KeyError as e:
  print("KeyError", e)
```

```python
data = data.sort_index()
data
```

```python
data['a': 'b']
```
