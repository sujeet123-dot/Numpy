```python
import pandas as pd
import numpy as np
```

```python
def make_df(cols, ind):
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

make_df('ABC', range(3))
```

```python
from re import template
class display(object):
  template = """<div style="float: left; padding: 10px; ">
  <p style='font-family: "Courier New", Courier, monospace'>{0}{1}
  """

  def __init__(self, *args):
    self.args = args

  def _repr_html_(self):
    return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                     for a in self.args)

  def __repr__(self):
    return '\n\n'.join(a + '\n' + repr(eval(a))
                       for a in self.args)
```

## Simple Concatenation with pd.concat

```python
ser1 = pd.Series(['A', 'B', 'C'], index=[1,2,3])
ser2 = pd.Series(['D', 'E', 'F'], index=[4,5,6])

pd.concat([ser1, ser2])
```

```python
df1 = make_df('AB', [1,2])
df2 = make_df('AB', [3,4])

display('df1', 'df2', 'pd.concat([df1, df2])')
```

ITS default behaviour is to concatenate row wise within the dataframe

### Duplicate indices

difference between **np.concatenate and pd.concate** is that pandas concatenation preserve indices even if the result will have duplicate indices

```python
x = make_df('AB', [0,1])
y = make_df('AB', [2,3])
y.index = x.index

display('x', 'y', 'pd.concat([x,y])')
```

```python
try:
  pd.concat([x,y], verify_integrity=True)
except ValueError as e:
  print("ValueError: ", e)
```

```python
# ignoring index

display('x', 'y', 'pd.concat([x,y], ignore_index=True)')
```

```python
# Multiindex keys

display('x', 'y', "pd.concat([x,y], keys=['x', 'y'])")
```

```python
df5 = make_df('ABC', [1,2])
df6 = make_df('BCD', [3,4])

display('df5', 'df6', 'pd.concat([df5, df6])')
```

```python
display('df5', 'df6', "pd.concat([df5, df6], join='inner')")
```

```python
pd.concat([df5, df6.reindex(df5.columns, axis=1)])
```
