A pivot table is similar operations that is commonly seen in spreadsheets and other programs that operate on tabular data.

Pivot table takes simple colums wise data as input and groups the entries onto two dimensional that provide multidimensional summarization of the data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex','class'])['survived'].aggregate('mean').unstack()
```

```python
# pivot table syntax
titanic.pivot_table('survived', index='sex', columns='class', aggfunc='mean')
```

```python
# Multilevel pivot tables

age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
# pd.qcut to automatically compute compile
```

```python
births = pd.read_csv('data/births.csv')
```

```python
births.head()
```

```python
births['decade'] = 10 * (births['year'] // 10)
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot styles
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year')
```

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
mu = quartiles[1]
sig = 0.74 * (quartiles[2] - quartiles[0])
```

```python
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
```

```python
births['day'] = births['day'].astype(int)
```

```python
births.index = pd.to_datetime(10000 * births.year +
                               100 * births.month +
                               births.day, format='%Y%m%d')
births['dayofweek'] = births.index.dayofweek
```

```python
import matplotlib as mpl

births.pivot_table('births', index='dayofweek',
                    columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day')
```

```python
births_by_date = births.pivot_table('births',
                                    [births.index.month, births.index.day])
births_by_date.head()
```

```python
from datetime import datetime

births_by_date.index = [datetime(2012, month, day)
for(month, day) in births_by_date.index]

births_by_date.plot(figsize=(12, 4))
plt.ylabel('average daily births')
```
