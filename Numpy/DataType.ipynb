## A single integer in python actually contain four spices
**ob_refcnt** : a refrence count that helps python silently handle memory allocation and deallocation

**ob_type**: which encodes the type of the variable

**ob_size**: which specifies the size of the following data mebers

**ob_digit**: which contains the actual integer value that we expect the python variable to represent

```python
L = list(range(10))
L
type(L[0])
L2 = [str(c) for c in L]
L2
```

we can create heterogeneous list because of python dynamic typing

```python
L3 = [True, "2", 3.0, 5]
[type(item)for item in L3]
```

## Fixed type Arrays in Python

python offer several different options for storing data in efficient, fixed type data buffers

```python
import array
L = list(range(10))
A = array.array('i', L)
A
```

Here, **' i '**  is a type code indicating the contents are integers

## Creating Arrays from Python lists

**np.array** is use to create arrays from python lists

unlike Python lists , Numpy arrays can only contain data of the same type. If the data type not matched Numpy will upcast them according to its type promotion rule

```python
import numpy as np
p = np.array([1,2,1,3,4])
q = np.array([3.15,2,5,6])
q
```

we can use explicitly data type by using of **dtype**

```python
np.array([1,2,3,4], dtype=np.float32)
```

Unlike python list, which are one dimensioanl sequence, Numpy array can be multidimensional

```python
np.array([range(i, i+3) for i in [2,4,6]])
```
