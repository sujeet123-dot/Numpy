```python
import numpy as np
```

```python
name = ['alice', 'bob', 'cathy', 'Doug']
age = [24, 45, 37, 19]
wight = [55.0, 85.5, 68.0, 61.5]
```

```python
x = np.zeros(4, dtype=int)
```

```python
# use compound data type for structured arrays

data = np.zeros(4, dtype={'names':('name', 'age', 'weight'),
                          'formats':('U10', 'i4', 'f8')})
print(data.dtype)
```

**U10** translate to "unicode string of maximum length 10"

**i4** translate to "4byte = 32 bits integer"

**f8** translate to "8 byte float"

```python
data['name'] = name
data['age'] = age
data['weight'] = wight
print(data)
```

```python
# Get all the name
data['name']
```

```python
# Get first row data
data[0]
```

```python
# get names where age is under 30
data[data['age'] < 30]['name']
```

## Structured array creation

```python
np.dtype({'names':('name', 'age', 'weight'),
         'formats':('U10', 'i4', 'f8')})
```

```python
# numerical type can be specified using python types or numpy dtype

np.dtype({'names':('name', 'age', 'weight'),
         'formats':((np.str_, 10), int, np.float32)})
```

```python
# compound type can also be specified as a list of tuples
np.dtype([('name', 'S10'), ('age', 'i4'), ('weight', 'f8')])
```

## More Advanced Compound Types

```python
tp = np.dtype([('id', 'i8'), ('mat', 'f8', (3,3))])
x = np.zeros(1, dtype=tp)
print(x[0])
print(x['mat'][0])
```
