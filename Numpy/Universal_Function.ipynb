## The Slowness of Loops
Python's default implementation (known as CPython) does some operations very slowly . This is partly due to dynamic, interpreted nature of the language; types are flexible so sequence of operations cannot be compiled down to efficient machine code like c

```python
import numpy as np
rng = np.random.default_rng(seed=1701)

def compute_reciprocals(values):
  output = np.empty(len(values))
  for i in range(len(values)):
    output[i] = 1.0 / values[i]
  return output

values = rng.integers(1, 10, size=5)
print(compute_reciprocals(values))
```

This might feel normal but for large data its taking some seconds to execute


```python
big_array = compute_reciprocals(rng.integers(1, 100, size=1000000))
%timeit compute_reciprocals(big_array)
```

## Array Arithmetic

```python
x = np.arange(4)

print("x     =", x)
print("x+5   =", x+5)
print("x-5   =", x-5)
print("x*5   =", x*5)
print("x/5   =", x/5)
print("x//5  =", x//5)
print("x**5  =", x**5)
print("x%5   =", x%5)
print("-x    =", -x)
```

## Trigonometric Functions

```python
theta = np.linspace(0, np.pi, 3)
```

```python
print("theta:   ", theta)
print("sin(theta):   ", np.sin(theta))
print("cos(theta):   ", np.cos(theta))
print("tan(theta):   ", np.tan(theta))
```

```python
# for inverse use arc prefix
x = [-1, 0, 1]
print("x         = ", x)
print("arcsin(x) = ", np.arcsin(x))
print("arccos(x) = ", np.arccos(x))
print("arctan(x) = ", np.arctan(x))
```

## Exponents and Logarithms

```python
x = [1,2,3]
print("x   =", x)
print("e^x =", np.exp(x))
print("2^x =", np.exp2(x))
print("3^x =", np.power(3., x))
```

```python
# inverse of exponential is logarithm
x = [1,2,4,10]
print("x        =", x)
print("ln(x)    =", np.log(x))
print("log2(x)  =", np.log2(x))
print("log10(x) =", np.log10(x))
```

## Specialized ufuncs

If you want to compute some obsecure mathematical function chances that it is implemented in **scipy.special**

```python
from scipy import special
```

```python
# Gamma function(generalized factorials) and related functions
x = [1, 5 ,10]
print("gamma(x)     =", special.gamma(x))
print("ln|gamma(x)| =", special.gammaln(x))
print("beta(x,2)    =", special.beta(x, 2))
```

```python
# error function (integral of Gaussian)
x = np.array([0,0.3,0.7,1.0])
print("erf(x)  =", special.erf(x))
print("erfc(x) =", special.erfc(x))
print("erfinv(x) =", special.erfinv(x))
```

## Specifying Output

For large calculations, it is sometime useful to specify the array where the result of the calculation will be stored

```python
x = np.arange(5)
y = np.empty(5)
np.multiply(x, 10, out=y)
print(y)
```

## Aggregations

For binary ufuncs, aggregations can be computed directly from the object.

A **reduce** repeatedly applies a given operation to the elements of an array until only a single result remains

```python
# reduce on the add ufunc return sum of all elements in the array
x = np.arange(1,6)
np.add.reduce(x)
```

```python
# reduce on multiply result in product of array element
np.multiply.reduce(x)
```

we can store all intermediate result using **accumulate**

```python
np.add.accumulate(x)
```

```python
np.multiply.accumulate(x)
```

## Outer Products

any ufunc can compute the output of all pairs of two different inputs using **outer** method

```python
x = np.arange(1,6)
np.multiply.outer(x,x)
```
