## Numpy Array attributes



```python
import numpy as np
rng = np.random.default_rng(seed=1701) # seed for reproduciblity

x1 = rng.integers(10, size=6) # ONE DIMENSIONAL
x2 = rng.integers(10, size=(3,4)) # Two dimensional
x3 = rng.integers(10, size=(3,4,5)) # Three dimensional

print("x3 dimensions", x3.ndim)
print("x3 shape", x3.shape)
print("x3 size", x3.size)
print("dtype:   ", x3.dtype)
print("itemsize:", x3.itemsize)
```

**ndim** : the number of dimensions

**shape**: the size of each dimension

**size**: the total size of array

**dtype**: type of each element

**itemsize**: size in byte of one element in array

## Array Slicing

syntax: x[start:stop:step]

### One Dimensional

```python
x1
```

```python
x1[:3] # first three element
```

```python
x1[3:] # elements after index 3
```

```python
x1[1:4] # middle subarray
```

```python
x1[::2] # every second element
```

```python
x1[1::2] # every second element, starting at index 1
```

### Multidimensional Subarrays

```python
x2
```

```python
x2[:2, :3] # first two rows and three column
```

```python
x2[::-1, ::-1] # all rows and column reversed
```

```python
x2[:, 0] # first column of x2
```

## Subarrays as No-Copy Views

Numpy array slices are returned as **views** rather than **copies** of the array data

```python
print(x2)
```

```python
x2_sub = x2[:2, :2]
print(x2_sub)
```

```python
x2_sub[0,0] = 99
print(x2_sub)
```

```python
print(x2)
```

## Copies of Arrays

**copy()** is used to copy

```python
x2_sub_copy = x2[:2,:2].copy()
print(x2_sub_copy)
```

```python
x2_sub_copy[0,0] = 42
print(x2_sub_copy)
```

```python
print(x2)
```

## Reshaping of Arrays

```python
grid = np.arange(1, 10).reshape(3,3)
print(grid)
```

**NOTE :** for this to work, the size of this initial array must match the size of the reshaped array, and in most cases the reshape method will return a no copy view of initial array

```python
x = np.array([1,2,3])
x.reshape((1,3)) # row vector via reshape
```

```python
x.reshape((3,1)) # column vector via reshape
```

A convenient shorthand is to use **np.newaxis**

```python
x[np.newaxis, :] # row vector via newaxis
```

```python
x[:, np.newaxis]
```

## Concatenation of Arrays

Contenation of two arrays in numpy by using **np.concatenate**, **np.vstack**, **np.hstack**

```python
x = np.array([1,2,3])
y = np.array([3,2,1])
np.concatenate([x,y])
```

```python
# Two dimensional
grid = np.array([[5,6,4],
                [8,7,6]])
np.concatenate([grid, grid])
```

```python
# concatenate along the second axis
np.concatenate([grid, grid], axis=1)
```

For working with array using mixed dimensions , use **np.vstack** (vertical stack), **np.hstack** (horizontal stack)

```python
np.vstack([x, grid])
```

```python
y = np.array([[99],
              [99]])
np.hstack([grid, y])
```

For higher dimensional arrays, **np.dstack** will stack arrays along third axis

## Splitting of Arrays

```python
x = [1,2,3,99,99,3,2,1]
x1, x2, x3 = np.split(x, [3,5])
print(x1,x2, x3)
```

**N** split points leads to **N + 1** SUBARRAYS

```python
grid = np.arange(16).reshape(4,4)
grid
```

```python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
```

```python
left, right = np,np.hsplit(grid, [2])
print(left)
print(right)
```
