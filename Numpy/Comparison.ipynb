## Comparison Operator

```python
x = np.array([1,6,2,5,3])

x > 3 # Greater than 3
```

```python
x < 3 # Less than
```

```python
x >= 3 # Greater than or equal
```

```python
x <= 3 # Less than or equal
```

```python
x != 3 # Not equal
```

```python
x == 3 # equal
```

## we can use Numpy ufuncs

+ (>)  = np.greater
+ (<)  = np.less
+ (<=) = np.less_equal
+ (>=) = np.greater_equal
+ (!=) = np.not_equal
+ (==) = np.equal

## Working with boolean

counting Entries

```python
rng = np.random.default_rng(seed=1071)
x = rng.integers(10, size=(3,3))
x
```

```python
# count value less than 6
np.count_nonzero(x < 6)
```

```python
# sum is another way to get value less than 6
np.sum(x < 6)
```

```python
# any value
np.any(x < 7)
```

```python
np.any(x > 7)
```

```python
np.all(x == 10)
```

## Bitwise Operator

+ (&) np.bitwise_and
+ (|) np.bitwise_or
+ (^) np.bitwise_xor
+ (~) np.bitwise_not

```python
np.sum((rainfall_mm > 10) & (rainfall_mm < 25))
```

```python
np.sum(~((rainfall_mm <= 10) & (rainfall_mm >= 25)))
```

```python
np.sum(~((rainfall_mm <= 10) | (rainfall_mm >= 25)))
```

**and** and **or** operates on the object as whole

**&** and **|** operates on the element within object

```python
A = np.array([1,0,1,0,1], dtype=bool)
B = np.array([1,1,1,0,1], dtype=bool)
A | B
```
